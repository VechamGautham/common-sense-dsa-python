# Unnecessary Recursive Calls 

```python
def maxi(array):

    if len(array) == 1 :
        return array[0]
    
    if array[0] > maxi(array[1:]):
        return array[0]
    else:
        return maxi(array[1:])

array = [1,2,3,4]

print(maxi(array))


```

time complexity O(2^N)

# Optimal Recurssive Call

```python
def max(array):
    print("recurssion")
    if len(array) == 1 :
        return array[0] 
    
    # calculate the max of the remainder of the array and store ir inside the variable rest_max
    rest_max = max(array[1:])

    # comparion of first number against this variable 
    if array[0] > rest_max :
        return array[0]
    else:
        return rest_max 


array = [1,4,2,3]

print(max(array))
```

time complexity O(N)

# Fibonacci Sequence

```python
def fib(n):
    print("recurssion")
    if n <= 1 :
        return n

    return fib(n-1) + fib(n-2)

print(fib(6))
```

time compleixty O(2^N)

# Dynamic Programming Through Memoization

```python
def fib(n,memo):
    print("recurssion")
    if n <= 1 :
        return n 
    
    if n not in memo :
        memo[n] = fib(n-2,memo) + fib(n-1,memo)
    
    return memo[n]

print(fib(n = 6,memo ={}))
```

time complexity O(2N) ==> O(N)
