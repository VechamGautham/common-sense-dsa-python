to create dir you could either use `os.mkdir("dir_name")` or `os.makedirs("dir_name")` the best about `os.makedirs("dir_name/sub_dir")` is the it allows us to make multiple dir and sub dir in one code which is not possibe with `os.makdir()`

`os.rmdir(dir_name)` remoes the dir_name that exist in the current directory you can also use `os.removedirs("dir_name")` to remove/delete dir and sub-dir in one go

````python
os.removedirs("OS-DEMO-2/SUB-DIR-1")
````

to rename a file or folder use `os.rename("orginal_name","name_you_want_to_change")`

````python
open("create_file.py","w").close()
print(os.listdir()) 
os.rename("create_file.py","renamed_create_file.py")
print(os.listdir())
````

`os.stat("file")` returns detailed info about the file like size, timestamps, and permissions.  
It's used to inspect metadata such as last modified time or file size.

````python
print(os.stat("renamed_create_file.py"))
print(os.stat("renamed_create_file.py").st_mtime)
````

importing `datetime` module to read the timestamp

````python
from datetime import datetime 

mod_time = os.stat("renamed_create_file.py").st_mtime

print(datetime.fromtimestamp(mod_time))
````

`os.walk("path")` generates all directories and files under a given path, walking top-down.  
It yields a tuple of (current_dir_path,dirnames,fileames) for each directory it visits.

````python
for dirpath, dirnames , filenames in os.walk("/Users/gauthamvecham/Desktop/dsa"):
    print("current path", dirpath)
    print("directories", dirnames)
    print("files", filenames)
````

`os.environ` is a dictionary-like object in Python that gives access to the operating system's environment variables.  
These variables store system-level settings like user info, file paths, or configuration values used by programs.

````python
print(os.environ.get("HOME"))
````

````python
file_path = os.environ.get("HOME") + "test.txt"

print(file_path)
````

you can combine path using `os.path.join(path1,path2)`

````python
file_path = os.path.join(os.environ.get("HOME"),"test.txt")

print(file_path)
````

### 🧭 Python `os.path` Functions – All in One Markdown

```python
import os

# os.path.exists(path)
# Checks if a file or directory exists
print(os.path.exists("example.txt"))
# Output: True

# os.path.basename(path)
# Returns the last part of the path (file or directory name)
print(os.path.basename("/home/user/example.txt"))
# Output: example.txt

# os.path.dirname(path)
# Returns the directory part of the path
print(os.path.dirname("/home/user/example.txt"))
# Output: /home/user

# os.path.split(path)
# Splits the path into a tuple: (directory, filename)
print(os.path.split("/home/user/example.txt"))
# Output: ('/home/user', 'example.txt')

# os.path.splitext(path)
# Splits the filename into (name, extension)
print(os.path.splitext("example.txt"))
# Output: ('example', '.txt')

# os.path.isdir(path)
# Checks if the path is a directory
print(os.path.isdir("/home/user"))
# Output: True

# os.path.isfile(path)
# Checks if the path is a file
print(os.path.isfile("example.txt"))
# Output: True
