# Understanding Recursion Step-by-Step (Using Factorial Example)

We’ll understand recursion using a simple factorial function and a fixed process:

1. Identify the base case  
2. Walk through the base case  
3. Identify the next-to-last case (the case just before the base case)  
4. Walk through the next-to-last case  
5. Repeat the process by moving one step outward each time until you reach the original input  

---

## The Factorial Code

Here is the factorial function we are analysing:

    def factorial(number):
        if number == 1 :
            return 1 
        else :
            return number * factorial(number - 1)
    
    
    print(factorial(number=3))

---

## 1. Find the Base Case

Look at the function and its two paths:

- The `if` statement: `if number == 1 : return 1`  
- The `else` statement: `return number * factorial(number - 1)`

The base case is the one that **does not** call the function again.  
So the base case is:

- If `number` is 1, return 1.

We can write this as:

- factorial(1) returns 1

Image:

![factorial(1) returns 1](./images/fact_1.png)

---

## 2. Next-to-Last Case (Number = 2)

The next-to-last case is the case just before the base case.  
Since the base case is `number = 1`, the next-to-last case is `number = 2`.

Evaluate factorial(2):

- factorial(2) goes to the `else` branch:
  - factorial(2) = 2 * factorial(1)

We already know from the base case:

- factorial(1) = 1

So:

- factorial(2) = 2 * 1 = 2

Store this information as well:

- factorial(2) returns 2

Image:

![factorial(2) returns 2, factorial(1) returns 1](./images/fact_2.png)

---

## 3. Move One Step Outward: factorial(3)

Now repeat the same process for the original call:

- factorial(3) = 3 * factorial(2)

We already know:

- factorial(2) = 2

So:

- factorial(3) = 3 * 2 = 6

So we have:

- factorial(3) returns 6

Image:

![factorial(3) returns 6, factorial(2) returns 2, factorial(1) returns 1](./images/fact_3.png)

---

## 4. Final Summary

Working from the base case upward:

- factorial(1) = 1  
- factorial(2) = 2 * factorial(1) = 2 * 1 = 2  
- factorial(3) = 3 * factorial(2) = 3 * 2 = 6  

The pattern for analysing recursion is:

1. Start with the base case and understand its exact return value.  
2. Take the next-to-last case, express it in terms of the base case, and evaluate it.  
3. Keep stepping outward until you reach the input you actually care about (here, 3).  

Once you’re comfortable with this pattern on factorial, you can apply the same approach to any recursive function.

```python
def factorial(number):
    if number == 1 :
        return 1 
    else :
        return number * factorial(number -1 )


print(factorial(number= 3))
```

```python
import os 

def print_subdirectories(directory_name):
    for filename in os.listdir(directory_name):
        path = os.path.join(directory_name,filename)

        if os.path.isdir(path):
            print(path)
            print_subdirectories(path)

print_subdirectories(".")
```
