## 54: Spiral Matrix

---

### üìù Description:

Given an `m x n` matrix, return all the elements of the matrix in **spiral order**.

---

### üî¢ Example:

```python
Input:
matrix = [
  [1, 2, 3],
  [4, 5, 6],
  [7, 8, 9]
]

Output: [1, 2, 3, 6, 9, 8, 7, 4, 5]

# Explanation:
# Start at the top-left corner, move right across the top row,
# then down the last column, then left across the bottom row,
# and finally up the first column. Continue inward in a spiral
# until all elements are visited.

````python
def spiral(matrix):
        m , n = len(matrix) , len(matrix[0])
        row , col = 0,0
        left,right,up,down = 0,1,2,3
        ans = []
        up_wall = 0
        right_wall = n 
        down_wall = m 
        left_wall = -1 

        direction = right
        while len(ans) != m*n :
            if direction == right :
                while col < right_wall :
                    ans.append(matrix[row][col])
                    col += 1 
                row,col = row+1,col -1 
                right_wall -= 1
                direction = down 
            elif direction == down :
                while row < down_wall:
                    ans.append(matrix[row][col])
                    row += 1 
                row,col = row -1 , col - 1 
                down_wall -= 1
                direction = left 
            elif direction == left :
                while col > left_wall:
                    ans.append(matrix[row][col])
                    col -= 1 
                row,col = row-1,col+1 
                left_wall += 1
                direction = up 
            elif direction == up :
                while row > up_wall:
                    ans.append(matrix[row][col])
                    row -= 1 
                row,col = row+1, col +1 
                up_wall += 1
                direction = right 
        return ans 
matrix = [
  [1, 2, 3],
  [4, 5, 6],
  [7, 8, 9]
]

print(spiral(matrix))


````


## Explanation
---

### üìù Problem Statement:

Given an `m x n` matrix, return all the elements of the matrix in **spiral order**.

You must start from the **top-left** corner and:
1. Move right until the end of the column.
2. Move down until the end of the row.
3. Move left until the beginning of the column.
4. Move up until the beginning of the row.
5. Repeat this process inward until **all elements are visited exactly once**.

---

### üéØ Objective:

Traverse the matrix in a **spiral** pattern without repeating any element.

---

### üöÄ Approach: Wall Boundaries + Direction Tracking

The best way to solve this problem is by simulating the traversal using:
- **Four imaginary walls**: `up_wall`, `right_wall`, `down_wall`, `left_wall`
- **Direction control**: right ‚Üí down ‚Üí left ‚Üí up

#### Key Idea:
- The "walls" act as **boundaries** that shrink inward after you finish a pass in one direction.
- Once you finish moving in a direction, you:
  - Adjust the respective wall so you don't revisit the same row/column.
  - Change direction clockwise.

---

### ‚öôÔ∏è Intuition:

1. **Set boundaries**:
   - `up_wall = 0` ‚Üí the topmost row we haven't traversed yet.
   - `right_wall = n` ‚Üí one step past the rightmost column index.
   - `down_wall = m` ‚Üí one step past the bottommost row index.
   - `left_wall = -1` ‚Üí one step before the leftmost column index.

2. **Initial position**:
   - Start at `(row=0, col=0)` moving to the **right**.

3. **Move in the current direction until hitting the wall**:
   - If moving **right**, stop before `col == right_wall`.
   - If moving **down**, stop before `row == down_wall`.
   - If moving **left**, stop before `col == left_wall`.
   - If moving **up**, stop before `row == up_wall`.

4. **After finishing one direction**:
   - Adjust the corresponding wall inward (so we don‚Äôt repeat those cells).
   - Change direction clockwise.

5. **Repeat** until the number of elements in `ans` equals `m * n`.

---

### üî¢ Example:

Input:
```python
matrix = [
  [1, 2, 3],
  [4, 5, 6],
  [7, 8, 9]
]
```

Step-by-step traversal:
	‚Ä¢	Right ‚Üí [1, 2, 3]
	‚Ä¢	Down ‚Üí [6, 9]
	‚Ä¢	Left ‚Üí [8, 7]
	‚Ä¢	Up ‚Üí [4]
	‚Ä¢	Right ‚Üí [5]

‚úÖ Final Output: [1, 2, 3, 6, 9, 8, 7, 4, 5]

‚∏ª

### ‚è±Ô∏è Time and Space Complexity:
	‚Ä¢	Time Complexity: O(m * n) ‚Äî each element is visited exactly once.
	‚Ä¢	Space Complexity: O(1) ‚Äî ignoring the output list, traversal is in-place with only variables for boundaries and direction.

## 867: Transpose Matrix 

---

### üìù Description:

Given a 2D integer array `matrix`, return **the transpose** of the matrix.  
The transpose of a matrix is obtained by flipping it over its main diagonal, switching the row and column indices of the matrix.

---

### üî¢ Example 1:

Input: matrix = [[1,2,3],[4,5,6],[7,8,9]]  
Output: [[1,4,7],[2,5,8],[3,6,9]]

**Explanation**:  
Original:        ‚Üí       Transposed:  
1 2 3            ‚Üí       1 4 7  
4 5 6            ‚Üí       2 5 8  
7 8 9            ‚Üí       3 6 9  

---

### üî¢ Example 2:

Input: matrix = [[1,2,3],[4,5,6]]  
Output: [[1,4],[2,5],[3,6]]

**Explanation**:  
Original:        ‚Üí       Transposed:  
1 2 3            ‚Üí       1 4  
4 5 6            ‚Üí       2 5  
                  ‚Üí       3 6  

---

### üö¶ Constraints:
- 1 <= matrix.length <= 1000  
- 1 <= matrix[0].length <= 1000  
- 1 <= matrix[i][j] <= 10^5  

````python
def transpose(matrix):
    transpose = []
    threshold = len(matrix[0])
    i = 0 
    while i < threshold:
        transpose_nested = []
        for nested in matrix :
            transpose_nested.append(nested[i])
        i += 1 
        transpose.append(transpose_nested)
    return transpose 
matrix = [[1,2,3],[4,5,6],[7,8,9]]

print(transpose(matrix))
````

# IN PLACE

````python
def transpose(matrix):
    n = len(matrix)

    for i in range(n):
        for j in range(i+1,n):
            matrix[i][j],matrix[j][i] = matrix[j][i],matrix[i][j]
    return matrix 

matrix = [[1,2,3],[4,5,6],[7,8,9]]

print(transpose(matrix))
````

# Explanation
### üìù Problem Statement:

Given an `m x n` matrix, return the **transpose** of the matrix.  
The transpose of a matrix is obtained by flipping it over its main diagonal, switching the row and column indices.

---

### üéØ Objective:

Convert **rows into columns** and **columns into rows**.

---

### üöÄ Approach: Index Mapping (Row ‚Üí Column)

The intuition is simple:
- For each **column index** in the original matrix, we extract all elements across rows.
- These collected elements form a **row in the transpose matrix**.

#### Key Idea:
- If `matrix` is `m x n`, then its transpose is `n x m`.  
- Example: `transpose[i][j] = matrix[j][i]`.

---

### üß† Intuition (exactly in the spirit of your explanation)

Think of the matrix as a list of **nested lists** (each nested list is a row).  
To transpose, you **don‚Äôt shuffle values inside a row**; instead, for each **column index** `i` in the original, you **pick that same index** from **every row** and those picked values together form **one row** in the transposed matrix.

- If original is `m √ó n`, the transpose is `n √ó m`.
- We therefore need **one new row per original column**.
- That‚Äôs why the **threshold** for the outer loop is `len(matrix[0])` (the number of columns in the original). It‚Äôs dynamic and works for any shape.

**Pick-the-same-index idea (your ‚Äúsame index across nested lists‚Äù)**  
- For `i = 0`, take `matrix[0][0]`, `matrix[1][0]`, `matrix[2][0]`, ‚Ä¶ ‚Üí that becomes the **first row** of the transpose.  
- For `i = 1`, take `matrix[0][1]`, `matrix[1][1]`, `matrix[2][1]`, ‚Ä¶ ‚Üí that becomes the **second row** of the transpose.  
- Keep increasing `i` until you‚Äôve covered all original columns.

This exactly mirrors your narrative: **‚Äútake index `i` from each nested list, gather them, and append as a new row.‚Äù**

---

### ‚öôÔ∏è Step-by-Step (mapped to your variables)

1. **Initialize**  
   - `transpose = []` (the final answer)  
   - `threshold = len(matrix[0])` (how many new rows we‚Äôll create = number of original columns)  
   - `i = 0` (the column index we‚Äôre currently extracting)

2. **Outer loop (`while i < threshold`)**  
   - Create `transpose_nested = []` ‚Üí this will become **one row** in the transpose.

3. **Inner loop (`for nested in matrix`)**  
   - Each `nested` is an original row.  
   - Append `nested[i]` to `transpose_nested`.  
   - After iterating all rows, `transpose_nested` has length `m` (one element from each original row).

4. **Append & Advance**  
   - `transpose.append(transpose_nested)`  
   - `i += 1` and repeat until `i == threshold`.

5. **Finish**  
   - Return `transpose`.

**Loop invariant clarity:**  
- After processing index `i`, the transpose has exactly `i + 1` rows.  
- Each new row‚Äôs length equals the original number of rows (`m`).

---

### üî¢ Example:

Input:
<pre>
matrix = [
[1, 2, 3],
[4, 5, 6],
[7, 8, 9]
]
</pre>

Step-by-step transformation:
- Pick index `0` ‚Üí [1, 4, 7]  
- Pick index `1` ‚Üí [2, 5, 8]  
- Pick index `2` ‚Üí [3, 6, 9]  

‚úÖ Final Transpose:
<pre>
[
[1, 4, 7],
[2, 5, 8],
[3, 6, 9]
]
</pre>

---

### ‚è±Ô∏è Time and Space Complexity:
- **Time Complexity:** O(m * n) ‚Äî every element is visited once.  
- **Space Complexity:** O(m * n) ‚Äî new transpose matrix is created. 



````python
def set_zero(matrix):
    indexes = []
    for i in range(len(matrix)):
        for j in range(len(matrix[0])):
            if matrix[i][j] == 0 :
                indexes.append([i,j])
    for i in range(len(matrix)):
        for j in range(len(matrix[0])):
            for index in indexes:
                matrix[index[0]][j] = 0 
                matrix[i][index[1]] = 0 
    
    return matrix 
print(set_zero(matrix=[[1,1,1],[1,0,1],[1,1,1]]))
print(set_zero(matrix=[[0,1,2,0],[3,4,5,2],[1,3,1,5]]))
        
````

## 48: Rotate Image

---

### üìù Description
You are given an `n x n` 2D matrix representing an image. Rotate the image by **90¬∞ clockwise** **in place** (modify the input directly; **no** extra 2D matrix).

---

### üî¢ Example 1
**Input:** `matrix = [[1,2,3],[4,5,6],[7,8,9]]`  
**Output:** `[[7,4,1],[8,5,2],[9,6,3]]`

**Explanation (Original ‚Üí Rotated 90¬∞):**
    1 2 3        7 4 1
    4 5 6   ‚Üí    8 5 2
    7 8 9        9 6 3

---

### üî¢ Example 2
**Input:** `matrix = [[5,1,9,11],[2,4,8,10],[13,3,6,7],[15,14,12,16]]`  
**Output:** `[[15,13,2,5],[14,3,4,1],[12,6,8,9],[16,7,10,11]]`

---

### üö¶ Constraints
- `n == matrix.length == matrix[i].length`
- `1 <= n <= 20`
- `-1000 <= matrix[i][j] <= 1000`




````python
def rotate_image(matrix):
        n = len(matrix)
        # transpose 
        for i in range(n):
            for j in range(i+1,n):
                matrix[i][j],matrix[j][i] = matrix[j][i],matrix[i][j]


        # reflection 

        for i in range(n):
            for j in range(n//2):
                matrix[i][j],matrix[i][n- j - 1] = matrix[i][n-j-1],matrix[i][j]
        return matrix  
    
print(rotate_image(matrix=[[5,1,9,11],[2,4,8,10],[13,3,6,7],[15,14,12,16]]))
````

## üåÄ Rotate Image ‚Äî Transpose + Horizontal Reflection (In-Place)

### Intuition
Rotate an `n √ó n` matrix 90¬∞ clockwise **without extra space** by doing two in-place passes:
1) **Transpose** the matrix (swap along the main diagonal), then  
2) **Reflect each row** horizontally (reverse each row).

This pair of operations is equivalent to a 90¬∞ clockwise rotation.

---

### How It Works

**1) Transpose (rows ‚Üî columns)**  
Swap `matrix[i][j]` with `matrix[j][i]` for all `j > i`.  
This flips the matrix over its main diagonal.

**2) Reflect each row (left ‚Üî right)**  
For each row `i`, swap symmetric elements:  
- `matrix[i][j]` ‚Üî `matrix[i][n - j - 1]` for `j` in `0 .. (n // 2) - 1`

Why `n // 2`?  
- You only need to process the **first half** of the columns to avoid double-swapping.  
- If `n` is **odd**, the middle column stays in place automatically.  
- If `n` is **even**, every column gets swapped exactly once.


---
<pre>
### Example Transformation
Original:
    5  1  9 11
    2  4  8 10
    13 3  6  7
    15 14 12 16

After transpose:
    5  2  13 15
    1  4  3  14
    9  8  6  12
    11 10 7  16

After reflecting rows (final rotated):
    15 13 2  5
    14 3  4  1
    12 6  8  9
    16 7 10 11
</pre>
---

### Complexity
- **Time:** O(n¬≤) (each element participates in at most a constant number of swaps)  
- **Space:** O(1) (in-place)

## 73: Set Matrix Zeroes

---

### üìù Description
You are given an `m x n` integer matrix `matrix`.  
If an element is `0`, set its **entire row and column** to `0`.  
You must do it **in place**, i.e., modify the input matrix directly ‚Äî **no** extra matrix allowed.

---

### üî¢ Example 1
**Input:**  
`matrix = [[1,1,1],[1,0,1],[1,1,1]]`  
**Output:**  
`[[1,0,1],[0,0,0],[1,0,1]]`

**Explanation:**  
The element at position `(1,1)` is `0`,  
so row `1` and column `1` are set to zeros.

---

### üî¢ Example 2
**Input:**  
`matrix = [[0,1,2,0],[3,4,5,2],[1,3,1,5]]`  
**Output:**  
`[[0,0,0,0],[0,4,5,0],[0,3,1,0]]`

**Explanation:**  
Zeros in positions `(0,0)` and `(0,3)`  
cause their entire rows and columns to be set to zero.

---

### üö¶ Constraints
- `m == matrix.length`
- `n == matrix[0].length`
- `1 <= m, n <= 200`
- `-2¬≥¬π <= matrix[i][j] <= 2¬≥¬π - 1`

---

### üí° Follow-up
- Can you do it **in-place** using **O(1)** extra space (constant space complexity)?

````python

def setmatrixzeroes(matrix):
    row_flag = col_flag = 0 
    rows = len(matrix)
    cols = len(matrix[0])

    for row in range(rows):
        for col in range(cols):
            if matrix[row][col] == 0 :
                if row == 0 :
                    row_flag = True
                if col == 0 :
                    col_flag = True 
                
                elif row != 0 and col != 0 :
                    matrix[0][col] = 0 
                    matrix[row][0] = 0 
    for row in range(1,rows):
        for col in range(1,cols):
            if matrix[row][0] == 0 or matrix[0][col] == 0 :
                matrix[row][col] = 0 
    
    if row_flag:
        matrix[0] = [0]*cols 
    
    if col_flag:
        for row in range(rows):
            matrix[row][0] = 0 
    
    return matrix 

print(setmatrixzeroes(matrix=[[1,1,1],[1,0,1],[1,1,1]]))

print(setmatrixzeroes(matrix=[[0,1,2,0],[3,4,5,2],[1,3,1,5]]))
````

### üîç Why the Second Loop Starts from (1, 1)

When we find zeros in the matrix during the **first pass**,  
we use the **first row** and **first column** to *mark* which rows and columns should be zeroed later.

- `matrix[0][col] = 0` ‚Üí means that **column** should become zero later  
- `matrix[row][0] = 0` ‚Üí means that **row** should become zero later  

So the **first row and first column act as storage** for these markers.

If we start the second loop from `(0, 0)`,  
we would accidentally overwrite these markers before using them ‚Äî  
and lose the information about which rows or columns should be zeroed.

That‚Äôs why the **second loop starts from row = 1 and col = 1** ‚Äî  
it skips the ‚Äúmarker area‚Äù and safely uses the stored information to zero out the rest of the matrix.

After this is done, we finally check the flags:
- If `row_flag` is True ‚Üí zero out the entire first row.  
- If `col_flag` is True ‚Üí zero out the entire first column.

So, in short:
1. First row/col ‚Üí used for marking.  
2. Skip them in the second pass.  
3. Handle them separately at the end.

## üí° Approach Explanation for ‚ÄúSet Matrix Zeroes‚Äù

---

### üß† Core Idea
The goal is to set entire rows and columns to zero when any element is zero ‚Äî but **without using extra space**.  
To achieve this, we cleverly use the **first row and first column** of the same matrix as markers.

We also maintain two boolean flags:
- **`row_flag`** ‚Üí to remember if the first row needs to be zeroed out.  
- **`col_flag`** ‚Üí to remember if the first column needs to be zeroed out.

---

### ü™ú Step-by-Step Explanation

#### **1Ô∏è‚É£ Identify zeros and mark corresponding row/column**
We start by scanning every element in the matrix.

- If a cell `matrix[row][col]` is **0**:
  - If it's in the **first row**, mark `row_flag = True`.
  - If it's in the **first column**, mark `col_flag = True`.
  - If it's **not** in the first row or first column,  
    we mark the corresponding row and column using:
    - `matrix[row][0] = 0` ‚Üí this marks that row should be zeroed.
    - `matrix[0][col] = 0` ‚Üí this marks that column should be zeroed.

So the first row and column act as ‚Äúnotebooks‚Äù that remember where zeros appeared.

---

#### **2Ô∏è‚É£ Convert the marked rows and columns to zeros**
Now that markings are done:
- We iterate through the matrix again (excluding the first row and first column).
- For each cell `(row, col)`:
  - If either its row or column was marked (i.e., if `matrix[row][0] == 0` or `matrix[0][col] == 0`),  
    then we set that cell to zero.

This ensures all affected rows and columns are properly zeroed based on the earlier markings.

---

#### **3Ô∏è‚É£ Handle the first row and first column separately**
Finally, we use the flags to handle the special case of the first row/column:
- If `row_flag` is **True**, we make the entire first row zero.
- If `col_flag` is **True**, we make the entire first column zero.

These two are handled separately because they were used for marking, so modifying them earlier would‚Äôve lost our reference markers.

---

### üß© Summary
- **Step 1:** Detect zeros and mark them in the first row/column.  
- **Step 2:** Use these markers to set the respective rows and columns to zero.  
- **Step 3:** Finally, zero out the first row and column if needed.

---

### ‚öôÔ∏è Complexity
- **Time Complexity:** `O(m √ó n)` ‚Äî every element is processed a constant number of times.  
- **Space Complexity:** `O(1)` ‚Äî only a few variables are used, no extra matrix.

---

### üí¨ Intuition Recap
Instead of maintaining separate arrays for rows and columns,  
we re-use the **matrix itself as the storage mechanism** for tracking zeros.  
The only extra memory comes from two flags ‚Äî that‚Äôs why it‚Äôs constant space.
