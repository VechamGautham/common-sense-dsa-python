## 54: Spiral Matrix

---

### üìù Description:

Given an `m x n` matrix, return all the elements of the matrix in **spiral order**.

---

### üî¢ Example:

```python
Input:
matrix = [
  [1, 2, 3],
  [4, 5, 6],
  [7, 8, 9]
]

Output: [1, 2, 3, 6, 9, 8, 7, 4, 5]

# Explanation:
# Start at the top-left corner, move right across the top row,
# then down the last column, then left across the bottom row,
# and finally up the first column. Continue inward in a spiral
# until all elements are visited.

````python
def spiral(matrix):
        m , n = len(matrix) , len(matrix[0])
        row , col = 0,0
        left,right,up,down = 0,1,2,3
        ans = []
        up_wall = 0
        right_wall = n 
        down_wall = m 
        left_wall = -1 

        direction = right
        while len(ans) != m*n :
            if direction == right :
                while col < right_wall :
                    ans.append(matrix[row][col])
                    col += 1 
                row,col = row+1,col -1 
                right_wall -= 1
                direction = down 
            elif direction == down :
                while row < down_wall:
                    ans.append(matrix[row][col])
                    row += 1 
                row,col = row -1 , col - 1 
                down_wall -= 1
                direction = left 
            elif direction == left :
                while col > left_wall:
                    ans.append(matrix[row][col])
                    col -= 1 
                row,col = row-1,col+1 
                left_wall += 1
                direction = up 
            elif direction == up :
                while row > up_wall:
                    ans.append(matrix[row][col])
                    row -= 1 
                row,col = row+1, col +1 
                up_wall += 1
                direction = right 
        return ans 
matrix = [
  [1, 2, 3],
  [4, 5, 6],
  [7, 8, 9]
]

print(spiral(matrix))


````


## Explanation
---

### üìù Problem Statement:

Given an `m x n` matrix, return all the elements of the matrix in **spiral order**.

You must start from the **top-left** corner and:
1. Move right until the end of the column.
2. Move down until the end of the row.
3. Move left until the beginning of the column.
4. Move up until the beginning of the row.
5. Repeat this process inward until **all elements are visited exactly once**.

---

### üéØ Objective:

Traverse the matrix in a **spiral** pattern without repeating any element.

---

### üöÄ Approach: Wall Boundaries + Direction Tracking

The best way to solve this problem is by simulating the traversal using:
- **Four imaginary walls**: `up_wall`, `right_wall`, `down_wall`, `left_wall`
- **Direction control**: right ‚Üí down ‚Üí left ‚Üí up

#### Key Idea:
- The "walls" act as **boundaries** that shrink inward after you finish a pass in one direction.
- Once you finish moving in a direction, you:
  - Adjust the respective wall so you don't revisit the same row/column.
  - Change direction clockwise.

---

### ‚öôÔ∏è Intuition:

1. **Set boundaries**:
   - `up_wall = 0` ‚Üí the topmost row we haven't traversed yet.
   - `right_wall = n` ‚Üí one step past the rightmost column index.
   - `down_wall = m` ‚Üí one step past the bottommost row index.
   - `left_wall = -1` ‚Üí one step before the leftmost column index.

2. **Initial position**:
   - Start at `(row=0, col=0)` moving to the **right**.

3. **Move in the current direction until hitting the wall**:
   - If moving **right**, stop before `col == right_wall`.
   - If moving **down**, stop before `row == down_wall`.
   - If moving **left**, stop before `col == left_wall`.
   - If moving **up**, stop before `row == up_wall`.

4. **After finishing one direction**:
   - Adjust the corresponding wall inward (so we don‚Äôt repeat those cells).
   - Change direction clockwise.

5. **Repeat** until the number of elements in `ans` equals `m * n`.

---

### üî¢ Example:

Input:
```python
matrix = [
  [1, 2, 3],
  [4, 5, 6],
  [7, 8, 9]
]
```

Step-by-step traversal:
	‚Ä¢	Right ‚Üí [1, 2, 3]
	‚Ä¢	Down ‚Üí [6, 9]
	‚Ä¢	Left ‚Üí [8, 7]
	‚Ä¢	Up ‚Üí [4]
	‚Ä¢	Right ‚Üí [5]

‚úÖ Final Output: [1, 2, 3, 6, 9, 8, 7, 4, 5]

‚∏ª

### ‚è±Ô∏è Time and Space Complexity:
	‚Ä¢	Time Complexity: O(m * n) ‚Äî each element is visited exactly once.
	‚Ä¢	Space Complexity: O(1) ‚Äî ignoring the output list, traversal is in-place with only variables for boundaries and direction.
