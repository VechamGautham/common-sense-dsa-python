### LeetCode: Min Cost Climbing Stairs  
ðŸ”— Link: https://leetcode.com/problems/min-cost-climbing-stairs/

You are given an integer array `cost` where `cost[i]` is the cost of stepping on the `i`-th stair. Once you pay the cost at a stair, you can either climb to the next stair (`i + 1`) or skip one stair and go to (`i + 2`).

You can start from either step `0` or step `1`.

#### Goal  
Return the **minimum cost** required to reach the **top of the floor** (just beyond the last index of the array).

#### Example 1  
Input: `cost = [10, 15, 20]`  
Output: `15`  
Explanation:  
- Start at index 1 (cost = 15), then directly move to the top.  
- Total cost = 15.

#### Example 2  
Input: `cost = [1, 100, 1, 1, 1, 100, 1, 1, 100, 1]`  
Output: `6`  
Explanation:  
- Take steps with costs: 1 (index 0) â†’ 1 (index 2) â†’ 1 (index 4) â†’ 1 (index 6) â†’ 1 (index 7)  
- Total cost = 1 + 1 + 1 + 1 + 1 = 5 (or similar minimal combination depending on path).

#### Constraints  
- `2 <= cost.length <= 1000`  
- `0 <= cost[i] <= 999`

# Brute Force Recursive

```python
count = 0 
def mincostfromm(i,cost):
    global count 
    count += 1 
    if i >= len(cost):
        return 0 
    
    return cost[i] + min(mincostfromm(i+1,cost),mincostfromm(i+2,cost))

def mincostclimbingstairs(cost):
    return min(mincostfromm(0,cost),mincostfromm(1,cost))

print(mincostclimbingstairs(cost=[1,100,1,1,100]))

print(count)
```

time complexity O(2^N) with brute force recursive

# Memoization method 

```python
count = 0 
def mincostfrom(i,cost,memo):
    global count 
    count += 1
    if i >= len(cost):
        return 0 
    
    if i not in memo :
        memo[i] = cost[i] + min(mincostfrom(i+1,cost,memo),mincostfrom(i+2,cost,memo))

    return memo[i]

def mincostclimbingstairs(cost,memo):

    return min(mincostfrom(0,cost,memo),mincostfrom(1,cost,memo))

print(mincostclimbingstairs(cost=[1,100,1,1,100],memo={}))
print(count)
```

time complexity O(2N) == > O(N) with memoization

# Bottom Up Approach 

```python

count = 0 
def mincostclimbingstairss(cost):
    n = len(cost)
    dp = [0] * (n+2) # dp[n] and dp[n+1] are 0 by default (base case)

    # fill from right to left
    for i in range(n-1,-1,-1):
        global count 
        count += 1
        dp[i] = cost[i] + min(dp[i+1],dp[i+2])
    
    return min(dp[0],dp[1])

cost = [1,100,1,1,100]

print(mincostclimbingstairss(cost))
print(count)

```

### LeetCode: Climbing Stairs  
ðŸ”— Link: https://leetcode.com/problems/climbing-stairs/

You are climbing a staircase with `n` steps. Each time you can either climb **1 step** or **2 steps**.

#### Goal  
Return the **number of distinct ways** to reach the top.

#### Example 1  
Input: `n = 2`  
Output: `2`  
Explanation:  
- 1 step + 1 step  
- 2 steps  

#### Example 2  
Input: `n = 3`  
Output: `3`  
Explanation:  
- 1 + 1 + 1  
- 1 + 2  
- 2 + 1  

#### Constraints  
- `1 <= n <= 45`

# Brute Force Recursive

```python
count = 0 
def climbingstairs(n):
    global count
    count += 1 
    if n <= 1 :
        return 1 
    return climbingstairs(n-1) + climbingstairs(n-2)

print(climbingstairs(10))
print(count)
```

time complexity for brute force recurssion O(2^N)

# Memoization method 

```python
count = 0 
def climbingstairs(n,memo):
    global count 
    count += 1 
    if n <= 1 :
        return 1 
    if n not in memo :
        memo[n] = climbingstairs(n-1,memo) + climbingstairs(n-2,memo)
    
    return memo[n]

print(climbingstairs(10,memo={}))
print(count)
```

time complexity for memoization method O(2N) == > O(N)

# Bottom - Up approach

```python
count = 0 
def climbingstairs_bu(n):
    dp = [0] * (n+1)
    dp[0] = 1 
    dp[1] = 1 

    for i in range(2,len(dp)):
        global count 
        count +=1 
        dp[i] = dp[i-1] + dp[i-2]
    
    return dp[n]

print(climbingstairs_bu(10))
print(count)
```

time complexity for bottom - Up approach is O(N)
